```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa
import librosa.display 
import seaborn as sns
from glob import glob
import IPython.display as ipd
from itertools import cycle
```

```python
sns.set_theme(style="white",palette=None)
color_pal = plt.rcParams['axes.prop_cycle'].by_key()['color']
color_cycle = cycle(plt.rcParams['axes.prop_cycle'].by_key()['color'])
```

```python
y,sr = librosa.load('p232_022.wav',sr=16000)
print(f'y: {y[:5]}')
print(f"Shape of y: {y.shape[0]}")
print("Sampling rate",sr)
```

```python
ipd.Audio('p232_022.wav')
```

```python
pd.Series(y).plot(figsize=(10,5),lw=1,title = "Raw audio sample",
                 color = color_pal[2])
plt.show()
```

```python
y_trimmed,_ = librosa.effects.trim(y,top_db=20)
```

```python
pd.Series(y_trimmed).plot(figsize=(10,5),lw=1,title = "Raw audio sample",
                 color = color_pal[1])
plt.show()
```

```python
pd.Series(y_trimmed[30000:31000]).plot(figsize=(10,5),lw=1,title = "Raw audio sample",
                 color = color_pal[3])
plt.show()
```
